### Exception Handling

##### Catching Exception

__try__ and __catch__ statement.
 
To handle errors (also known as exceptions) in Python, you can use the __tryâ€¦except__ statements. These statements tell Python what to do when an exception is encountered. 

try:
    statements # statements that can raise exceptions
except:
    statements # statements that will be executed to handle exceptions
    
Example:

```python
age=int(input('Enter your age: '))

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

```python
try:
    age=int(input('Enter your age: '))
except:
    print ('You have entered an invalid value.')

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

##### Catching Specific Exceptions


```python
try:
    age=int(input('Enter your age: '))
except Exception as e:
    print ('You have entered an invalid value. Error occured: '+ str(e))

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

```python
try:
    age=int(input('Enter your age: '))
except ValueError as e:
    print ('You have entered an invalid value. Error occured: '+ str(e))

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

```python
try:
    age=int(input('Enter your age: '))
    x_value = 10/age
    print(x_value)
except ValueError as e:
    print ('You have entered an invalid value. Error occured: '+ str(e))

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

##### Catching multiple Exceptions using multiple blocks

```python
try:
    age=int(input('Enter your age: '))
    x_value = 10/age
    print(x_value)
except ValueError as e:
    print ('You have entered an invalid value. Value Error occured: '+ str(e))
except ZeroDivisionError as e:
    print ('You have entered an invalid value. Zero Division Error occured: '+ str(e))

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

##### Catching multiple Exceptions using single block

```python
try:
    age=int(input('Enter your age: '))
    x_value = 10/age
    print(x_value)
except (ValueError,ZeroDivisionError) as e:
    print ('You have entered an invalid value. Error occured: '+ str(e))

if age <= 21:
    print('You are not allowed to enter, you are too young.')
else:
    print('Welcome, you are old enough.')
```

```python
class B(Exception):
    pass

class C(B):
    pass

class D(C):
    pass

for cls in [B, C, D]:
    try:
        raise cls()
    except D:
        print("D")
    except C:
        print("C")
    except B:
        print("B")
```

##### try with else clause

In some situations, you might want to run a certain block of code if the code block inside try ran without any errors. For these cases, you can use the optional else keyword with the try statement.

Note: Exceptions in the else clause are not handled by the preceding except clauses.

```python
try:
    num = int(input("Enter a number: "))
    assert num % 2 == 0
except:
    print("Not an even number!")
else:
    reciprocal = 1/num
    print(reciprocal)
```

##### try...finally
The try statement in Python can have an optional finally clause. This clause is executed no matter what, and is generally used to release external resources.

```python
try:
    f = open("errors_tut1.py",'r')
    f.write("hello")
except Exception as e:
    print("Something went wrong: " + str(e))
finally:
    print("Closing the file")
    f.close()
    
```

##### Using them all : try ...except...else..finally

```python
def divide(x, y): 
    try: 
        # Floor Division : Gives only Fractional 
        # Part as Answer 
        result = x // y 
    except ZeroDivisionError: 
        print("Sorry ! You are dividing by zero ") 
    else:
        print("Yeah ! Your answer is :", result) 
    finally:  
        # this block is always executed   
        # regardless of exception generation.  
        print('This is always executed')   
 
# Look at parameters and note the working of Program 
a = int(input("Enter a number: "))
b = int(input("Enter b number: "))
divide(a, b) 
```

##### Creating Custom Exceptions

Users can define custom exceptions by creating a new class. This exception class has to be derived, either directly or indirectly, from the built-in Exception class. Most of the built-in exceptions are also derived from this class.



```python
class ErrorValueInCode(Exception):   
    def __init__(self, data):      
        message = f"Got an incorrect value {data}"
        super().__init__(message,123)     
      
try:      
    raise ErrorValueInCode(2000)      
except ErrorValueInCode as ae:      
    print("Received error:", ae) 
    print(ae.args)
    print(ae.with_traceback())
#     print("Data:", ae.data)  
raise ErrorInCode(2000)
```
